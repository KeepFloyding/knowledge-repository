```r
# Read data
df = read.csv("raw_data.csv")

head(df)
```

```r
y = df$X0

nx = length(y)
x=seq(0, 10, length.out=nx)
plot(x,y)
```

```r
alpha = 0.05
cutoff = qt(1 - (alpha / nx) / 2, nx-2-1)/6
sum(abs(rstudent(lm(y~poly(x,3)))) > cutoff)
```

```r
tf = abs(rstudent(lm(y~poly(x,3)))) > cutoff
plot(x,y)
points(x[!tf],y[!tf], col="green", pch="+")
points(x[tf],y[tf], col="red", pch="+")
```

```r
alpha = 0.01
cutoff = qt(1 - (alpha / nx) / 2, nx-2-1)
sum(abs(rstudent(lm(y~poly(x,3)))) > cutoff)
```
