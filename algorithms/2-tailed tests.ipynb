# Understanding Statistical Tests

As far as I understand, there are several parts to a distribution function

* CDF: Cumulative Distribution Function; is the probability that X will take a value less than or equal to x
* PPF: Percent Point Function; specifies the value of the random variable such that the probability of the variable being less than or equal to that value equals the given probability.




```python
from scipy.stats import t as student_dist
import numpy as np
import matplotlib.pyplot as plt

dof=3

x = np.linspace(-5,5)
y = student_dist.cdf(x, df=dof)

plt.plot(x,y)
```

```python
from scipy.stats import norm
# 68–95–99.7
alpha=0.045
norm.ppf(alpha/2)
```

```python
norm.sf(5)
```

```python

dof=3

x = np.linspace(0,1)
y = student_dist.ppf(x, df=dof)

plt.plot(x,y)
```
